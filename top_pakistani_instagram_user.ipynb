```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="darkgrid")
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('Top Instragram.csv')
df.head()
```

```python
df = df[['Name', 'Category', 'Followers']]
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
def followers_to_float(x):
    if type(x) == float or type(x) == int:
        return x
    if 'K' in x:
        if len(x) > 1:
            return float(x.replace('K', '')) * 1000
        return 1000.0
    if 'M' in x:
        if len(x) > 1:
            return float(x.replace('M', '')) * 1000000
        return 1000000.0
    
df['Followers'] = df['Followers'].apply(followers_to_float)
df.head()
```

```python
df.isnull().sum()
```

```python
df['Category'].unique()
```

```python
# Changing category "Singer/Musician" to "Singer"
df.replace("Singer/Musician", "Singer", inplace=True)
df['Category'].unique()
```

```python
plt.figure(figsize=(10, 10))
dfCat = df.groupby(by=['Category']).sum()
subs = list(dfCat['Followers'])
names = list(df['Category'].unique())
colors = ['#baff9e', '#ff9c59', '#85dae6', '#e6859c', '#e69585', '#c1e685', '#87e685', '#e685dc', '#edabcc', '#c8e8c5']
plt.pie(subs, labels=names, autopct='%1.1f%%', colors=colors)
plt.show()
```
